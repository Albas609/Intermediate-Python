Selesaikan masalah missing value dengan cara berikut:

A) Drop kolom-kolom yang mengandung missing value lebih dari 50% keseluruhan data.

B) Isi missing value pada kolom country dan children dengan data yang paling banyak muncul.

C) Isi missing value pada kolom agent dengan mean.

```python
#Drop kolom-kolom yang mengandung missing value lebih dari 50% keseluruhan data.

data_hilang = hotel.isna()
data_hilang.head()
```

```python
x = data_hilang.sum()
x
```

```python
x / len(hotel)
```

```python
x.dropna(axis=0, how='any', inplace=True)
x
```

```python
# Isi missing value pada kolom country dan children dengan data yang paling banyak muncul.

hotel['country'] = hotel['country'].fillna(hotel['country'].mode()[0])
hotel['children'] = hotel['children'].fillna(hotel['children'].mode()[0])
```

```python
hotel
```

```python
# Isi missing value pada kolom agent dengan mean.

tengah = hotel.fillna(('agent'), (hotel.mean()))
```
